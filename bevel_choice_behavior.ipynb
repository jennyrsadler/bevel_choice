```python
#Goal = calculate probability of each shape as the sum of number of previous rewards/sum of number of previous reinforcers
#running matched/total reinforcers = A prob of correct
#running mismatched/total reinforcers = B prob of being correct

#df.loc[(df.choice == 'A') & (df.outcome == 'reward'), 'a_rewarded'] = '1' 
#df.loc[(df.choice == 'A') & (df.outcome == 'punish'), 'a_rewarded'] = '0' 
#df.loc[(df.choice == 'B') & (df.outcome == 'reward'), 'b_rewarded'] = '1' 
#df.loc[(df.choice == 'B') & (df.outcome == 'punish'), 'b_rewarded'] = '0' 
#df.loc[(df.img_pair == 'AB') & (df.outcome != 'Miss'), 'ab_reinforcer_count'] = '1' 
#print(df)

#Split df into AB
df1 = df[df['img_pair'] == 'AB']
df1
```

```python
cond = (df1.matched == 'matched')
#print(cond)
df1['matched_running'] = cond.cumsum().where(cond, 0).astype(int)
df1

cond1 = (df1.matched == 'mismatched')
#print(cond)
df1['mismatched_running'] = cond1.cumsum().where(cond1, 0).astype(int)
df1

cond2 = (df1.matched != 'Miss')
#print(cond)
df1['reinforcer_running'] = cond2.cumsum().where(cond2, 0).astype(int)
df1

df1['a_corr_prob'] = df1.matched_running / df1.reinforcer_running
df1['b_corr_prob'] = df1.mismatched_running / df1.reinforcer_running
df1
```
