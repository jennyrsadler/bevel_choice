```python
import os
import glob
import pandas as pd
import numpy as np
import pdb

basepath='/Users/jennygilbert/Documents/bevel_choice/by_participant_txtfiles'
os.chdir(basepath)
```

```python
df = pd.read_csv(os.path.join(basepath,'01.txt'), sep='\t', header = None, names = ['ID','run','img_pair','choice','side','outcome','matched','RT'])
df1 = df[df['img_pair'] == 'AB'] 
df1
```

```python
count = df1.choice.count()
list = np.arange(0, count)
list
```

```python
df1["matched_count"] = ""
df1["running_matched_count"] = ""

for i in list: 
    if df1.iloc[i, 6] == "matched":
        df1.iloc[i, 8] = 1
    
    elif df1.iloc[i, 6] == "mismatched":
        df1.iloc [i, 8] = 0
    else: 
        df1.iloc[i, 8] = 0
df1        
```

```python
df1["running_matched_count"] = df1["matched_count"].cumsum()
```

```python
df1
```

```python
cond2 = (df1.matched != 'Miss')
df1['reinforcer_running'] = cond2.cumsum().where(cond2, 999).astype(int)

df1 = df1.replace(999,np.NaN)

df1['a_learnedprob_correct'] = df1.running_matched_count / df1.reinforcer_running
df1['b_learnedprob_correct'] = 1 - df1.a_learnedprob_correct
df1
```
